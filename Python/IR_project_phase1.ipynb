```python
# User-Tag matrix . Perform similarity measure between users - Pearson/Cosine similarity
import heapq

def cosine_similarity(user1,user2):
    #print(user1,user2)
    vec1 = dummy_tag_df.loc[user1].values
    vec2 = dummy_tag_df.loc[user2].values
    cos_dist = np.dot(vec1,vec2) / (np.linalg.norm(vec1)*np.linalg.norm(vec2))
    #print(cos_dist)
    return cos_dist



## Compute and store for all users to avoid recomputing
# cosine_similarity_store = {}
# pearson_similarity_store = {}  ## will do in the relavant section
# for i in users:
#     for j in users:
#         if i != j:
#             #print(i,j)
#             cosine_similarity_store[(i,j)] = cosine_similarity(i,j)
            
#             cosine_similarity_store[(j,i)] = cosine_similarity_store[(i,j)]

#get top k similar users


def get_similar_user(target_user):
    max_k_scores =  [(0,0) for i in range(5)]      
    heapq.heapify(max_k_scores)
    for comparing in users:
        if comparing != target_user :
            heapq.heappushpop(max_k_scores,(cosine_similarity(comparing,target_user),comparing)) 
            heapq.heapify(max_k_scores)
    return max_k_scores

target_user_id = 143   #### change this value to get recos for differenct users
trempa = get_similar_user(target_user_id)
print(trempa)
print("\n Target user history \n", sorted(user_city_dict[target_user_id]))
print("\n\n Recommended user \n\n",sorted(user_city_dict[trempa[4][1]]))
print("\n",sorted(user_city_dict[trempa[3][1]]))
print("\n",sorted(user_city_dict[trempa[2][1]]))
#print("\n",sorted(user_city_dict[trempa[1][1]]))

#print("\n",sorted(user_city_dict[trempa[0][1]]))

```
