```python
def user_vector(df,user_id):
    final_data_frame = ['User','Numeric_Stay_Length','Category_Trip_Expenses','Circumambience_id']
    rel_frame = df.loc[df['User'] == user_id]
    rec_stay = rel_frame['Numeric_Stay_Length'].value_counts().idxmax()
    rec_trip = rel_frame['Category_Trip_Expenses'].value_counts().idxmax()
    rec_circum = rel_frame['Circumambience_id'].value_counts().idxmax()
    vector = pd.DataFrame(columns=final_data_frame)
    vector.loc[0] = [user_id,rec_stay,rec_trip,rec_circum]
    return vector.values
```
